```python
#Calculadora IMC
peso = float(input("ingresar el peso"))
altura = float(input("Ingresa tu altura"))

imc = peso / (altura ** 2)
print (f' Tu IMC es: (imc:2f)')

if imc < 15.5:
  print("bajo normal")

elif imc < 20:
   print("peso normal")

elif imc < 25:
  print("Sobre peso")

else:

  print("Eres un kilo mortal")
```

```python
#operacion con listas
#Quintana Maria Jose
frutas = ['manzana', 'pera', 'uva', 'mango']

print(frutas[0]) # mostrar la primera fruta
print(frutas[-1]) # mostrar ultima fruta

frutas[1]= 'naranja' # notificacion de posición 1
frutas.append('kiwi') # Agregar al final
frutas.insert(1, 'fresa') # Insertar en posición 1
frutas.remove('uva') # Eliminar la 1ra.
frutas.pop(0) # Eliminar y retornar el indice
```

```python
#lista con numeros
#Quinana Maria Jose

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]
print('Original:', numeros)

# Ordenr de menor a mayor
numeros.sort()
print('ordenada:0', numeros)

# Ordenar de mayor a menor
numeros.reverse()
print('invertida:', numeros)
```

```python
#lista con numeros
#Quinana Maria Jose

total = sum(numeros)
promedio = sum(numeros) / len(numeros)
print(f"La suma total es: {total}")
print(f"El promedio es: {promedio}")
```
